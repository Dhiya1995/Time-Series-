```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset
url = '/content/AirPassengers.csv'
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Perform decomposition
result = seasonal_decompose(data['#Passengers'], model='multiplicative', period=12)
```

```python

# 1. ðŸ“ˆ Original (Observed) Data
plt.figure(figsize=(10, 4))
plt.plot(data['#Passengers'], label='Observed')
plt.title('Original Time Series (Observed)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# 2. ðŸ“‰ Trend Component
plt.figure(figsize=(10, 4))
plt.plot(result.trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.xlabel('Date')
plt.ylabel('Trend')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# 3. ðŸ” Seasonal Component
plt.figure(figsize=(10, 4))
plt.plot(result.seasonal, label='Seasonality', color='green')
plt.title('Seasonal Component')
plt.xlabel('Date')
plt.ylabel('Seasonality')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# 4. ðŸ” Residual Component
plt.figure(figsize=(10, 4))
plt.plot(result.resid, label='Residuals', color='red')
plt.title('Residual Component')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.legend()
plt.tight_layout()
plt.show()

```
